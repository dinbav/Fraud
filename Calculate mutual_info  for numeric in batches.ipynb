```
import os
def start_timer():
  (utime, stime) = os.times()[0:2]
  start_time= utime+stime
  return start_time

def end_timer(start_time):
  (utime, stime) = os.times()[0:2]
  end_time= utime+stime
  time_to_string= "%05.5f seconds" % (end_time-start_time)
  return time_to_string
```

```
df_numeric_columns= list(df_numeric.columns) 
```

```
def calculate_mi(x,y):
  print('Start calculating mutual info for columns:' + str(list(x.columns)))
  start_time=start_timer()
  feature_scores = mutual_info_classif(x, y)
  print('Mutual info calculation took '+end_timer(start_time))
  for i in range(len(x.columns)):
    print(x.columns[i]+','+str(feature_scores[i]))
```

```
def batch(iterable, n=1):
    l = len(iterable)
    for ndx in range(0, l, n):
        yield iterable[ndx:min(ndx + n, l)]
```

```
for x in batch(df_numeric_columns, 5):
    calculate_mi(df_numeric[x],y)
```
